```python
import ast

def main():
    # Choose an operation
    operation = input("Enter the operation (addition, subtraction, multiplication, division, or power): ").lower()

    # Enter a list
    list_input = input("Enter a list of 5 numbers (comma-separated): ")
    try:
        input_list = [float(num) for num in list_input.split(',')]
        if len(input_list) != 5:
            raise ValueError("Invalid list size.")
    except ValueError:
        print("Please enter a valid list.")
        return

    # Enter a set
    set_input = input("Enter a set of 5 numbers (comma-separated): ")
    try:
        input_set = {float(num) for num in set_input.split(',')}
        if len(input_set) != 5:
            raise ValueError("Invalid set size.")
    except ValueError:
        print("Please enter a valid set.")
        return

    # Enter a dictionary
    dict_input = input("Enter a dictionary with 5 key-value pairs (comma-separated, e.g., key1:value1,key2:value2): ")
    try:
        input_dict = dict(pair.split(':') for pair in dict_input.split(','))
        input_dict = {key: float(value) for key, value in input_dict.items()}
        if len(input_dict) != 5:
            raise ValueError("Invalid dictionary size.")
    except (ValueError, KeyError):
        print("Please enter a valid dictionary.")
        return

    # Check if division by zero
    if operation == "division" and (0 in input_set or 0 in input_dict.values()):
        print("Cannot divide by zero.")
        return

    # Perform the selected operation
    if operation == "addition":
        result = sum(input_list) + sum(input_set) + sum(input_dict.values())
    elif operation == "subtraction":
        result = sum(input_list) - sum(input_set) - sum(input_dict.values())
    elif operation == "multiplication":
        result = 1
        for num in input_list:
            result *= num
        for num in input_set:
            result *= num
        for num in input_dict.values():
            result *= num
    elif operation == "division":
        result = sum(input_list) / sum(input_set) / sum(input_dict.values())
    elif operation == "power":
        result = sum(input_list) ** sum(input_set) ** sum(input_dict.values())
    else:
        print("Invalid operation.")
        return

    # Print the input values, operation, and result
    print(f"List: {input_list}")
    print(f"Set: {input_set}")
    print(f"Dictionary: {input_dict}")
    print(f"Operation: {operation}")
    print(f"Result: {result}")

# Call the main function
if __name__ == "__main__":
    main()
```
